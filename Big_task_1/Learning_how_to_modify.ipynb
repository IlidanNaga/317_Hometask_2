```python
from sklearn.datasets import fetch_mldata
from random import shuffle
from random import seed
from matplotlib import pyplot as plt
```

```python
mnist = fetch_mldata("MNIST-original")
data = mnist.data / 255.0
target = mnist.target.astype("int0")


# i'd take smaller subsets for faster calculations
seed(1024)
indixes_list = np.arange(70000)
shuffle(indixes_list)
sub_data = data[indixes_list[:3500]]
sub_target = target[indixes_list[:3500]]
```

```python
plt.imshow(sub_data[0].reshape(28, 28))
```

```python
rot_data = skimage.transform.rotate(sub_data.reshape(3500, 28, 28), 15).astype(np.float)
```

```python
sub_data = sub_data.reshape(3500, 28, 28)
rot_data = np.empty(sub_data.shape)
```

```python
for i in range(3500):
    rot_data[i] = skimage.transform.rotate(sub_data[i], -15)
```
