```python
import numpy as np
import pandas as pd
import re

from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

def check_metrics(true_val, pred_val):
    print("Accuracy: ", accuracy_score(true_val, pred_val))
    print("Roc_Auc: ", roc_auc_score(true_val, pred_val))
    print("Precision: ", precision_score(true_val, pred_val), " - процент правильно токсичных")
    print("Recall: ", recall_score(true_val, pred_val), " - процент выбранных токсичных")

data = pd.read_csv("toxic_train.csv")

test_data = pd.read_csv("toxic_test.csv")


train_texts = np.array(data['comment_text'])
test_texts = np.array(test_data['comment_text'])

for enum in range(train_texts.__len__()):
    train_texts[enum] = train_texts[enum].lower()
    train_texts[enum] = re.sub("[^0-9a-zA-Z]+", ' ', train_texts[enum])
    train_texts[enum] =  re.sub(' +', ' ', train_texts[enum])


    
test_texts = np.array(test_data['comment_text'])
for enum in range(test_texts.__len__()):
    test_texts[enum] = test_texts[enum].lower()
    test_texts[enum] = re.sub("[^0-9a-zA-Z]+", ' ', test_texts[enum])
    test_texts[enum] =  re.sub(' +', ' ', test_texts[enum])


train_target = np.array(data['is_toxic']).astype(int) * 2 - 1 
test_target = np.array(test_data['is_toxic']).astype(int) * 2 - 1


```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect = CountVectorizer(min_df = 100, stop_words = "english")
```

```python
train_matrix = vect.fit_transform(train_texts)
```

```python
test_matrix = vect.transform(test_texts)
```

```python
from optimization import SGDClassifier
```

```python
classifier = SGDClassifier(batch_size = int(0.7 * train_texts.__len__()), step_alpha = 2.1, l2_coef = 0)
```

```python
%time classifier.fit(train_matrix, train_target)
```

```python
check_metrics(test_target, classifier.predict(test_matrix))
```

О, качество чуть улучшилось

Рассмотрим с различными размерами batch_size относительно полного размера выборки

```python
results = []
holds = []
for i in range(1, 21):
    classifier = SGDClassifier(batch_size = int(i / 20 * train_texts.__len__()), step_alpha = 2.6, l2_coef = 0)
    holds.append(classifier.fit(train_matrix, train_target, trace = True))
    
    results.append(classifier.predict(test_matrix))
```

```python
for i in range(20):
    print(accuracy_score(test_target, results[i]))
```

Значит, оптиумумом можно пока считать 0.45 от размера тренировочной выборки

Для оптимального значения batch_size выберем step_alpha

```python
results_alpha = []
holds_alpha = []
for i in range(1, 41):
    classifier = SGDClassifier(batch_size = int(0.45 * train_texts.__len__()),max_iter = 100, step_alpha = i/10, l2_coef = 0)
    holds_alpha.append(classifier.fit(train_matrix, train_target, trace = True))
    
    results_alpha.append(classifier.predict(test_matrix))
```

```python
al = []
for i in range(40):
    al.append(accuracy_score(test_target, results_alpha[i]))
```

Рассмотрим step_beta:

```python
results_beta = []
holds_beta = []
for i in range(1, 11):
    classifier = SGDClassifier(batch_size = int(0.45 * train_texts.__len__()), step_alpha = 1.6, step_beta = i/10, max_iter = 100, l2_coef = 0)
    holds_beta.append(classifier.fit(train_matrix, train_target, trace = True))
    
    results_beta.append(classifier.predict(test_matrix))
```
